```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df_jujuy = pd.readcsv('./datasets/GC_UNJujuy_process.txt', sep='\t')
#df_palermo = pd.readcsv('./datasets/GC_UniPalermo_process.txt', sep='\t')
df_jujuy = pd.read_csv('/content/GC_UNJujuy_process.txt', sep='\t')
df_palermo = pd.read_csv('/content/GC_UniPalermo_process.txt', sep='\t')
```

```python
df_jujuy = df_jujuy[df_jujuy['age'] <= 90]
```

```python
df_aux = df_palermo.groupby('gender').count().reset_index()

plt.figure(figsize=(6, 7))
sns.barplot(data=df_aux, x='gender', y='university')
plt.xlabel('Genero')
plt.ylabel('Cantidad de personas')
plt.title('Cantidad de personas por genero en la Universidad De Palermo')
plt.show()
```

```python
df_aux = df_jujuy.groupby('gender').count().reset_index()

plt.figure(figsize=(6, 7))
sns.barplot(data=df_aux, x='gender', y='university')
plt.xlabel('Genero')
plt.ylabel('Cantidad de personas')
plt.title('Cantidad de personas por genero en la Universidad Nacional De Jujy')
plt.show()
```

```python
df_aux = df_palermo.groupby('career').count().reset_index().sort_values('university', ascending=False)

plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

plt.figure(figsize=(13, 10))
sns.barplot(data=df_aux, x='university', y='career')
plt.xlabel('Cantidad de personas')
plt.ylabel('Careras')
plt.title('Cantidad de personas por carrera - Universidad De Palermo')
plt.show()
```

```python
df_aux = df_jujuy.groupby('career').count().reset_index().sort_values('university', ascending=False)

plt.figure(figsize=(13, 10))
sns.barplot(data=df_aux, x='university', y='career')
plt.xlabel('Cantidad de personas')
plt.ylabel('Careras')
plt.title('Cantidad de personas por carrera - Universidad Nacional De Jujuy')
plt.show()
```

```python
plt.figure(figsize=(14, 14))
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
sns.countplot(data=df_palermo, y='career', hue='gender')
plt.xlabel('Cantidad de personas')
plt.ylabel('Careras')
plt.title('Cantidad de personas por carera por genero - Universidad De Palermo')
plt.show()
```

```python
plt.figure(figsize=(14, 14))
sns.countplot(data=df_jujuy, y='career', hue='gender')
plt.xlabel('Cantidad de personas')
plt.ylabel('Careras')
plt.title('Cantidad de personas por carera por genero - Nacional De Jujuy por genero')
plt.show()
```

## Edades


```python
df_aux = df_palermo.groupby('age').count().reset_index()

plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

plt.figure(figsize=(19, 8))
plt.bar(data=df_aux, x='age', height='university')
plt.title('Cantidad de personas por edad - Universidad De Palermo')
plt.xlabel('Edad')
plt.ylabel('Cantidad de personas')
plt.show()

plt.figure(figsize=(19, 5))
plt.boxplot(data=df_aux, x='age', vert=False)
plt.xlabel('Edad')
plt.ylabel('Boxplot')
plt.show()
```

```python
df_aux = df_jujuy.groupby('age').count().reset_index()

plt.figure(figsize=(19, 8))
plt.bar(data=df_aux, x='age', height='university')
plt.title('Cantidad de personas por edad - Universidad Nacional De Jujuy')
plt.xlabel('Edad')
plt.ylabel('Cantidad de personas')
plt.show()

plt.figure(figsize=(19, 5))
plt.boxplot(data=df_aux, x='age', vert=False)
plt.xlabel('Edad')
plt.ylabel('Boxplot')
plt.show()
```

```python
df_aux = df_palermo.groupby('career').mean().reset_index()

plt.figure(figsize=(13, 10))
sns.barplot(data=df_aux, x='age', y='career')
plt.xlabel('Edad promedio')
plt.ylabel('Careras')
plt.title('Edad promedio por carrera - Universidad De Palermo')
plt.show()

df_aux_max_min = df_palermo.groupby('career').max().reset_index()
df_aux_max_min['age_min'] = list(df_palermo.groupby('career').min()['age'])

df_aux_max_min.plot.barh(x = 'career', y = ['age', 'age_min'], figsize=(13,13), title="Edad minima y maxima por carrera - Universidad De Palermo", xlabel="Carreras", ylabel="Edad minimas y maximas")
```

```python
df_aux = df_jujuy.groupby('career').mean().reset_index()

plt.figure(figsize=(13, 10))
sns.barplot(data=df_aux, x='age', y='career')
plt.xlabel('Edad promedio')
plt.ylabel('Careras')
plt.title('Edad promedio por carrera - Universidad Nacional De Jujuy ')
plt.show()

df_aux_max_min = df_jujuy.groupby('career').max().reset_index()
df_aux_max_min['age_min'] = list(df_jujuy.groupby('career').min()['age'])

df_aux_max_min.plot.barh(x = 'career', y = ['age', 'age_min'], figsize=(13,13), title="Edad minima y maxima por carrera - Universidad Nacional De Jujuy ", xlabel="Carreras", ylabel="Edad minimas y maximas")
```

```python
plt.figure(figsize=(13, 10))
sns.countplot(data=df_palermo, y='location',  order=df_palermo['location'].value_counts().index[0:30])
plt.xlabel('Cantidad de personas')
plt.ylabel('Localidades')
plt.title('Cantidad de personas por localidad que van a la universidad - Universidad De Palermo')
plt.show()

plt.figure(figsize=(13, 10))
sns.countplot(data=df_palermo, y='location', hue='gender', order=df_palermo['location'].value_counts().index[0:30])
plt.xlabel('Cantidad de personas por genero')
plt.ylabel('Localidades')
plt.title('Cantidad de personas de cada genero por localidad que van a la universidad - Universidad De Palermo')
plt.show()
```

```python
plt.figure(figsize=(13, 10))
sns.countplot(data=df_jujuy, y='location',  order=df_jujuy['location'].value_counts().index[0:30])
plt.xlabel('Cantidad de personas')
plt.ylabel('Localidades')
plt.title('Cantidad de personas por localidad que van a la universidad - Universidad Nacional De Jujuy')
plt.show()

plt.figure(figsize=(13, 10))
sns.countplot(data=df_jujuy, y='location', hue='gender', order=df_jujuy['location'].value_counts().index[0:30])
plt.xlabel('Cantidad de personas por genero')
plt.ylabel('Localidades')
plt.title('Cantidad de personas de cada genero por localidad que van a la universidad - Universidad Nacional De Jujuy')
plt.show()
```

```python
df_aux = pd.concat([df_jujuy, df_palermo])
df_aux = df_aux.groupby('university').count().reset_index()

plt.figure(figsize=(13, 10))
plt.barh(data=df_aux ,y='university', width='inscription_date')
plt.xlabel('Cantidad de personas incriptas')
plt.ylabel('Universidades')
plt.title('Cantidad de personas inscriptas por universidad')
plt.show()
```

```python
df_aux = pd.concat([df_jujuy, df_palermo])
df_aux['inscription_date'] = pd.to_datetime(df_aux['inscription_date'], format='%Y-%m-%d')
df_aux = df_aux.groupby(['university','inscription_date']).count().reset_index()

plt.figure(figsize=(13, 10))
sns.lineplot(data=df_aux, x='inscription_date', y='career', hue='university') 
plt.xlabel('Fecha de isncripcion')
plt.ylabel('Cantidad de inscriptos')
plt.title('Cantidad de personas inscriptas por universidad en el tiempo')
plt.show()
```

```python
df_auxp = df_palermo.reset_index()
df_auxp['inscription_date'] = pd.to_datetime(df_auxp['inscription_date'], format='%Y-%m-%d')
df_auxp = df_auxp.set_index('inscription_date').resample('M').count()
df_auxp['university'] = "universidad de palermo"

df_auxj = df_jujuy.reset_index()
df_auxj['inscription_date'] = pd.to_datetime(df_auxj['inscription_date'], format='%Y-%m-%d')
df_auxj = df_auxj.set_index('inscription_date').resample('M').count()
df_auxj['university'] = "universidad nacional de jujuy"

df_aux = pd.concat([df_auxj, df_auxp]).reset_index()

plt.figure(figsize=(13, 10))
sns.lineplot(data=df_aux, x='inscription_date', y='career', hue='university') 
plt.xlabel('Fecha de isncripcion')
plt.ylabel('Cantidad de inscriptos')
plt.title('Cantidad de personas inscriptas por universidad por mes')
plt.show()
```

## Correlacion entre variables

```python
df_aux = pd.concat([df_palermo, df_jujuy])
df_aux = df_aux[['university','career','inscription_date','gender','age','postal_code','location']]
df_aux['career']=df_aux['career'].astype('category').cat.codes
df_aux['gender']=df_aux['gender'].astype('category').cat.codes
df_aux['location']=df_aux['location'].astype('category').cat.codes
df_aux['university']=df_aux['university'].astype('category').cat.codes

plt.figure(figsize=(10, 7))
sns.heatmap(df_aux.corr(), annot=True)
plt.title('Correlacion entre variables entre todas las universidades')
plt.show()
```

```python
df_aux = df_palermo.reset_index()
df_aux = df_aux[['career','inscription_date','gender','age','postal_code','location']]
df_aux['career']=df_aux['career'].astype('category').cat.codes
df_aux['gender']=df_aux['gender'].astype('category').cat.codes
df_aux['location']=df_aux['location'].astype('category').cat.codes

plt.figure(figsize=(10, 7))
sns.heatmap(df_aux.corr(), annot=True)
plt.title('Correlacion entre variables Universidad De Palermo')
plt.show()
```

```python
df_aux = df_jujuy.reset_index()
df_aux = df_aux[['career','inscription_date','gender','age','postal_code','location']]
df_aux['career']=df_aux['career'].astype('category').cat.codes
df_aux['gender']=df_aux['gender'].astype('category').cat.codes
df_aux['location']=df_aux['location'].astype('category').cat.codes

plt.figure(figsize=(10, 7))
sns.heatmap(df_aux.corr(), annot=True)
plt.title('Correlacion entre variables Universidad Nacional De Jujuy')
plt.show()
```
